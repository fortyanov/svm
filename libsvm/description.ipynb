```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, r2_score
from utils.visualisation import decision_boundary_plot
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

Подгружаем датасет с ирисами Фишера

```python
X1, y1 = load_iris(return_X_y=True, as_frame=True)
X1_train, X1_test, y1_train, y1_test = train_test_split(X1.values, y1.values, random_state=0)
print(X1, y1, sep='\n')
```

Работа с моделью:
* инициализация
* обучение 
* предсказание

```python
sk_svc = SVC(random_state=0, gamma='auto')
sk_svc.fit(X1_train, y1_train)
sk_svc_pred_res = sk_svc.predict(X1_test)
```

Отображение результатов
* ожидаемые результаты
* результаты предсказания
* меткрика точность
* метрика R2
* отображение

```python
sk_svc_accuracy = accuracy_score(y1_test, sk_svc_pred_res)

print(f'expected res:\t{y1_test}')
print(f'sk SVС res:\t{sk_svc_pred_res}')

print(f'sk SVC accuracy: {sk_svc_accuracy:}')
print(sk_svc_pred_res)

sk_svc_r2 = r2_score(y1_test, sk_svc_pred_res)
print(f'LinearSVM r2:\t\t{sk_svc_r2}')

decision_boundary_plot(X1, y1, X1_train, y1_train, sk_svc, feature_indexes=[2, 3], title='SVC surface')
```
